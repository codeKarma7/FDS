```python
#importing required libraries
import pandas as pd #for dataframes
import matplotlib.pyplot as plt #for plotting
import seaborn as sns #for plotting
import numpy as np

df = pd.read_csv("employees.csv")
df
```

```python
df.shape
```

```python
### 1. Data cleaning and Preprocessing starts from here
df.isnull().sum() #out of 500 data, data ma deko jati chai null xa
```

```python
# DropNone values for columns(EmployeeID, Name, JoiningDate )

df.dropna(subset = ['EmployeeID','Name','JoiningDate'],inplace=True)
df.shape #kati ota data baki xa check garna
df.info()
df
```

```python
#checking if any column of salary has null value

df['Salary'].isnull().sum()

#calculating mean of the salary
mean_salary = df['Salary'].mean()
print(mean_salary)

#in salary column jaha none xa tyaha mean_salary replace garne
df['Salary']=df['Salary'].fillna(mean_salary)
df
```

```python
#again ensuring that if a salary has missing value or not
df['Salary'].isnull().any()
```

```python
#checking if any column of age has missing value or not
missing_age = df['Age'].isnull().sum()
print(f"Missing value in 'Age': {missing_age}")
```

```python
#calculating mode of the age
mode_age = df['Age'].mode()[0]
print(f"Mode of 'Age': {mode_age}")

#replacing mode_age in missing values in table
df['Age']=df['Age'].fillna(mode_age)
df
```

```python
#checking if any age has missing value

df['Age'].isnull().any()
```

```python
#checking if department has missing value
df[' Department'].isnull().sum()
```

```python
#calculating mode for department
mode_depart = df[' Department'].mode()[0]
print(f"Mode of 'Department' : {mode_depart}")

#now replacing missing values with mode result in department section
df[' Department'] = df[' Department'].fillna(mode_depart)
df
```

```python
#reassuring if the department has missing values
df[' Department'].isnull().any()
```

```python
#checking number if missing values in Experience(years)
missing_experience = df['Experience (Years)'].isnull().sum()
print(f"Misssing values in 'Experience(years)' : {missing_experience}")
```

```python
#calculating median of the experience(year)
median_experience=df['Experience (Years)'].median()
print(f"Median of 'Experience(year)' : {median_experience}")

#replacing missing value in experience(year) with result of median_experience
df['Experience (Years)'] = df['Experience (Years)'].fillna(median_experience)
df
```

```python
#reassuring if there is a missing value in experience(year)
df['Experience (Years)'].isnull().any()
```

```python
#converting dataypes into best choice

#checking the datatypes of features
print(df.dtypes)
```

```python
#converting the feature into suitable datatype

df['EmployeeID'] = df['EmployeeID'].astype(int)
df['JoiningDate'] = pd.to_datetime(df['JoiningDate'])
df['Age'] = df['Age'].astype(int)
df['Experience (Years)'] = df['Experience (Years)'].astype(int)
df['Last Promotion Date'] = pd.to_datetime(df['Last Promotion Date'])

#again checking if the conversion of data type takes place 
print(df.dtypes)


#Data cleaning and preprocessing ends here
```

### 6. Performance and Salary Analysis : 

1. Is there any significant difference in the average Salary for different Departments?
 Which department has the highest average salary?

Ans : 

```python
#calculating average salary for each department
avg_salary_by_dept = df.groupby(' Department')['Salary'].mean()
print(f"Average salary by department : {avg_salary_by_dept}")

print("\n")

#calculating difference for each department's salary
salary_range = df['Salary'].max() - df['Salary'].min()
print(f"\nRange of salaries: {salary_range:.2f}")

#calculating the department with highest avg salary 
highest_salary_department = df.loc[df['Salary'].idxmax()]
print(f"The department with the highest average salary is {highest_salary_department[' Department']} with an average salary of {highest_salary_department['Salary']:.2f}.")
```

Ans : Hence,the significant difference in the average salary for different Departments is 3927.00. And,the department that has higher salary is Operation Department as per the data.

2. Isthere a significant difference in Salary based on Gender (e.g., do males and
 females have a similar salary distribution)

Ans:

```python
#calculating average salaryby gender

# Calculate the average salary for each gender
male_avg_salary = df[df['Gender '] == 'Male']['Salary'].mean()
female_avg_salary = df[df['Gender '] == 'Female']['Salary'].mean()

# Output the results
print(f"Average salary for males: {male_avg_salary:.2f}")
print(f"Average salary for females: {female_avg_salary:.2f}")

#calculating difference
diff_salary = (male_avg_salary) - (female_avg_salary)
print(f"The difference in Salary by gender : {diff_salary}")
```

Ans : As per the data we can clearly see the difference between the salary by gender is only 200 by average. So,here is no significant difference in salary based on gender.

3.  What isthe relationship between Performance Rating and Salary? Does
 performance correlate with pay?

Ans : 

```python
# calculating the correlation coefficient between performance rating and salary
correlation = np.corrcoef(df['Performance Rating'], df['Salary'])[0, 1]

#result of correlation
print(f"The correlation coefficient between Performance Rating and Salary is: {correlation:.2f}")

# Interpretation of the correlation
if correlation > 0.5:
    print("There is a strong positive correlation between Performance Rating and Salary.")
elif correlation > 0:
    print("There is a moderate positive correlation between Performance Rating and Salary.")
elif correlation == 0:
    print("There is no linear correlation between Performance Rating and Salary.")
elif correlation < -0.5:
    print("There is a strong negative correlation between Performance Rating and Salary.")
else:
    print("There is a moderate negative correlation between Performance Rating and Salary.")
```

Ans : Hence, as per the data There is a strong positive correlation between Performance Rating and Salary.
