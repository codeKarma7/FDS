```python
#load data from csv file
import pandas as pd
df = pd.read_csv("routine.csv")
df
```

```python
#Data cleaning

df.drop(columns=['Timestamp'], inplace= True)
df.drop(columns=['If you have breakfast, what do you eat?'], inplace=True)
df.drop(columns = ['Do you think your morning routine influences your productivity?'], inplace =True)

#renaming
df.rename(columns = {"How many hours of sleep do you get per night on average?" : "Sleep duration"},inplace =True)
df.rename(columns = {"Do you have Breakfast?" : "Breakfast"},inplace = True )
df.rename(columns = {"Time you wake-up" : "Wakeup-Time"},inplace = True)
df.rename(columns = {"How do you rate your energy level throughout the day?" : "Rate your energy level"}, inplace = True)
df.rename(columns = {"What is the first thing you do after waking up?" : "First thing to do"} ,inplace= True)
df
```

```python
#checking missing value
df.isna().sum()
```

```python
#data analysis

df['WakeUpCategory'] = pd.to_datetime(df['Wakeup-Time'], errors='coerce').dt.hour.apply(
    lambda h: 'Early Riser' if h < 7 else 'Late Riser' if h > 8 else 'Moderate Riser'
)

# Group and count data by wake-up category
grouped_data = df['WakeUpCategory'].value_counts()
grouped_data
```
