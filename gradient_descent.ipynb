Gradient Descent Algorithm and Learning curve for the linear function y=4+3x

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate the cost
def compute_cost(X, y, theta):
    m = len(y)
    predictions = X.dot(theta)
    errors = predictions - y
    cost = (1 / (2 * m)) * np.sum(errors ** 2)
    return cost

# Gradient descent implementation
def gradient_descent(X, y, theta, learning_rate, iterations):
    m = len(y)
    cost_history = []

    for i in range(iterations):
        predictions = X.dot(theta)
        errors = predictions - y
        gradient = (1 / m) * X.T.dot(errors)
        theta -= learning_rate * gradient

        # Save the cost for each iteration
        cost_history.append(compute_cost(X, y, theta))

    return theta, cost_history

# Generate synthetic data for demonstration
np.random.seed(0)
X = 2 * np.random.rand(100, 1)  # 100 samples, 1 feature
y = X**2  # Linear relation without noise

# Add a column of ones to X for the bias term
X_b = np.c_[np.ones((100, 1)), X]

# Initialize parameters
initial_theta = np.zeros((2, 1))  # 2 parameters (bias + slope)
learning_rate = 0.1
iterations = 100

# Perform gradient descent
final_theta, cost_history = gradient_descent(X_b, y, initial_theta, learning_rate, iterations)

# Plot the learning curve
plt.figure(figsize=(8, 6))
plt.plot(range(iterations), cost_history, 'b-', linewidth=2)
plt.title("Learning Curve", fontsize=14)
plt.xlabel("Number of Iterations", fontsize=12)
plt.ylabel("Cost", fontsize=12)
plt.grid(True)
plt.show()
```

For J(theta)=theta1 **2 +3 theta2

```python
import numpy as np
import matplotlib.pyplot as plt

# New cost function
def compute_cost(theta):
    cost = theta[0]**2 + 3 * theta[1]
    return cost

# Gradient descent for the new cost function
def gradient_descent(theta, learning_rate, iterations):
    cost_history = []

    for i in range(iterations):
        # Compute gradients manually for the new cost function
        grad_theta1 = 2 * theta[0]  # Derivative of theta1^2
        grad_theta2 = 3  # Derivative of 3 * theta2
        gradient = np.array([grad_theta1, grad_theta2])

        # Update theta
        theta -= learning_rate * gradient

        # Save the cost for each iteration
        cost_history.append(compute_cost(theta))

    return theta, cost_history

# Initialize parameters
initial_theta = np.zeros(2)  # [theta1, theta2]
learning_rate = 0.1
iterations = 1000

# Perform gradient descent
final_theta, cost_history = gradient_descent(initial_theta, learning_rate, iterations)

# Plot the learning curve
plt.figure(figsize=(8, 6))
plt.plot(range(iterations), cost_history, 'b-', linewidth=2)
plt.title("Learning Curve", fontsize=14)
plt.xlabel("Number of Iterations", fontsize=12)
plt.ylabel("Cost", fontsize=12)
plt.grid(True)
plt.show()
```
