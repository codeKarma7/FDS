```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("synthetic_sales_data.csv")
df
```

```python
#checking how many features and observations are there
df.shape
```

```python
df.isnull().sum()
```

```python
#renaming the feature that having white space
df.columns=df.columns.str.strip()
df.isnull().sum()
```

```python
#removing the data that doesn't have customerId,customerName,transactionId
df.dropna(subset=['Transaction_ID','Customer_ID','Customer_Name','Date'],inplace=True)
df
```

```python
#checking if observations has dropeed or not
df.shape
```

```python
#asking for the file info
df.info()
```

```python
#again checking null value of the feature
df.isnull().sum()
```

```python
#for null value in gender replacing with not specified
df['Customer_Gender']=df['Customer_Gender'].fillna("Not Specified")
df['Customer_Region']=df['Customer_Region'].fillna("Not Specified")
df['Discount_Amount']=df['Discount_Amount'].fillna(0)
df['Store_Location']=df['Store_Location'].fillna("Not Specified")
df['Product_Rating']=df['Product_Rating'].fillna("Not Rated")

mode_category=df['Category'].mode()[0]
print(f"Mode for category:{mode_category}")
df['Category']=df['Category'].fillna(mode_category)


median_price=df['Price'].median()
print(f"Median for price:{median_price}")
df['Price']=df['Price'].fillna(median_price)


mode_quantity=df['Quantity'].mode()[0]
print(f"Mode for quantity:{mode_quantity}")
df['Quantity']=df['Quantity'].fillna(mode_quantity)


mode_payment=df['Payment_Method'].mode()[0]
print(f"Mode for payment method:{mode_payment}")
df['Payment_Method']=df['Payment_Method'].fillna(mode_payment)


mode_online=df['Online/Offline'].mode()[0]
print(f"Mode for online/offline:{mode_online}")
df['Online/Offline']=df['Online/Offline'].fillna(mode_online)


median_productrating = df['Category'].median()
print(f"Median for product rating :{median_productrating}")
df['Price']=df['Price'].fillna(median_productrating)

df
```

```python
#checking datatypes of feastures
df.dtypes
```

```python
#converting into suitable datatypes
df['Date']=pd.to_datetime(df['Date'])
df['Quantity']=df['Quantity'].astype(int)
df.dtypes
```

```python
#checking number of unique value for different features
df.nunique()
```

```python
#euta data lai different type le lekheko xa ki xna check gareko
df['Category'].unique()
```

```python
#replacing category with only one unique category
df['Category']=df['Category'].replace({'Furniture':'Furnitures','electronic':'Electronics','ELECTRONIC':'Electronics'})
df
```

```python
#euta data lai different type le lekheko xa ki xna check gareko
df['Customer_Region'].unique()
df['Customer_Region']=df['Customer_Region'].replace({'WST':'West','Est':'EAST','NORTH':'North','south':'South','EAST':'East','WEST':'West'})
df['Customer_Region'].unique()
```

```python
#euta data lai different type le lekheko xa ki xna check gareko
df['Store_Location'].unique()
```

```python
#calculating total amount
df['Total_Amount']=(df['Price'] * df['Quantity']) - df['Discount_Amount']
df
```

```python
#calculating discount percentage
df['Discount_Percentage']=(df['Discount_Amount']/(df['Price'] * df['Quantity'])) * 100
df['Discount_Percentage']=df['Discount_Percentage'].round(2)
df
```

```python
#adding new feature CLV(Customer Lifetime Value)
customerID_counts=df['Customer_ID'].value_counts()
print(customerID_counts)
```

```python
#aggregate transactions by Customer_ID
clv_df = df.groupby('Customer_ID')['Total_Amount'].sum().reset_index()
clv_df.rename(columns={'Total_Amount': 'CLV'}, inplace=True)

#normalize CLV using min-max scaling to rate from 0 t0 10
min_clv = clv_df['CLV'].min()
max_clv = clv_df['CLV'].max()
clv_df['CLV_Scaled'] = ((clv_df['CLV'] - min_clv) / (max_clv - min_clv)) * 10

print(clv_df)
```

```python
#EDA Starts 

#calculating average spend for each customer in each category(tspc = total spend per customer, aspc = average spend per customer)
tspc_category = df.groupby(['Customer_ID', 'Category'])['Total_Amount'].sum().reset_index()
aspc_category = tspc_category.groupby(['Customer_ID', 'Category'])['Total_Amount'].mean().reset_index()
print(aspc_category)
```

```python
#calculating frequency of purchase 

#counting the number of transactiond for each customer_id and store_lovation
purchase_freq = df.groupby(['Customer_ID', 'Store_Location']).size().reset_index(name='Frequency')
print(purchase_freq)
```

```python
##preferred payment method 

#dentifying the most frequently used payment method for each customer( pp = preferred payment)

pp_method = df.groupby('Customer_ID')['Payment_Method'].agg(lambda x: x.mode()[0]).reset_index()
print(pp_method)
```

```python
##online/offline preference

#determining the percentage of online vs offline purchases for each customer
onoff_preference = df.groupby(['Customer_ID', 'Online/Offline']).size().unstack(fill_value=0)
onoff_preference_percentage = onoff_preference.div(onoff_preference.sum(axis=1), axis=0) * 100
print(onoff_preference_percentage)
```

```python
##product pupularity

#counting the number of times each Product_Type is purchased and rank them.
prod_popularity= df['Category'].value_counts().reset_index(name='Purchase_Count')
prod_popularity = prod_popularity.sort_values(by='Purchase_Count', ascending=False)
prod_popularity['Rank'] = np.arange(1, len(prod_popularity)+1)
print(prod_popularity)
```

```python
##Data visualization starts here


#list of numeric features
numeric_features = ['Price', 'Discount_Amount', 'Quantity', 'Total_Amount', 'Product_Rating']

#skewness
print("Skewness for each numeric feature:")
for feature in numeric_features:
    print(f"{feature}: {skew(df[feature].dropna())}")

#correlation matrix
correlation_matrix = df[numeric_features].corr()

#displaying the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Plotting histograms for visual skewness check
plt.figure(figsize=(10, 8))
for i, feature in enumerate(numeric_features, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df[feature], kde=True)
    plt.title(f"Distribution of {feature}")
plt.tight_layout()
plt.show()
```

```python
# List of categorical features (add the names of your categorical columns)
categorical_features = ['Category', 'Customer_Gender', 'Payment_Method', 'Store_Location', 'Online/Offline']  # Modify as per your data

# Plotting count plots for categorical features
plt.figure(figsize=(10, 8))
for i, feature in enumerate(categorical_features, 1):
    plt.subplot(2, 3, i)
    sns.countplot(x=df[feature])
    plt.title(f"Distribution of {feature}")
plt.tight_layout()
plt.show()
```
